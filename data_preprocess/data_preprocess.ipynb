```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
def remove_outliers(data, m=5):
    data = np.array(data)
    valid_idx = np.where(abs(data - np.mean(data)) < m * np.std(data))
    data = data[valid_idx]
    return data
```

```python
def get_df_all(year_list=[2002, 2005, 2008, 2011, 2014, 2017]):
    df_all = pd.DataFrame()
    df_city_info = pd.read_csv('../data/df_city_info.csv')
    # print(df_city_info.shape)
    # df_city_info.head()
    for year in year_list:
        df_one_year = pd.read_csv('../data/df_{}.csv'.format(year))
        df_one_year['year'] = year
        df_all = pd.concat([df_all, df_one_year], axis=0)
    df_all = df_all.reset_index(drop=True)
    df_all = df_all[df_all['UI'] > 0].reset_index(drop=True)
    df_all['EVI'] = preprocessing.minmax_scale(remove_outliers(df_all['EVI'], m=5))
    df_all = pd.merge(left=df_all, right=df_city_info[['city_id', 'cz']], on='city_id', how='left')
    return df_all
```

```python
year_list = [2002, 2005, 2008, 2011, 2014, 2017]
df_all = get_df_all()
```
