```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
```

```python
df_corr_wi = pd.read_csv('./table_partial_corr_Wi.csv')

plt.rc('font', family='Times New Roman', size=10)
linewidth = 0.4
plt.rcParams['axes.linewidth'] = linewidth
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['xtick.major.width'] = linewidth
plt.rcParams['ytick.major.size'] = 3
plt.rcParams['ytick.major.width'] = linewidth
plt.figure(figsize=(7.5, 2.3), dpi=150)

plt_id = 0
for cz_area in ['Global', 'Tropical', 'Temperate', 'Cold', 'Arid']:
    plt_id += 1
    df_corr_wi_ = df_corr_wi[df_corr_wi['y_name'] == 'wi_mean'].reset_index(drop=True)
    df_corr_wi_ = df_corr_wi_[df_corr_wi_['cz'] == cz_area].reset_index(drop=True)
    
    plt.subplot(1, 5, plt_id)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    df_corr_wi_ = df_corr_wi_.iloc[::-1]
    plt.barh(df_corr_wi_['x_name'], df_corr_wi_['r'], color=df_corr_wi_['color'], height=0.68)
    if plt_id > 1:
        plt.yticks([0,1,2,3,4,5],())
    plt.xlim(-0.55, 0.55)
    plt.xticks([-0.5, 0, 0.5])
    cz_name = cz_area
plt.tight_layout()
plt.show()
```

```python
df_impor_wi = pd.read_csv('./table_variable_importances_Wi_by_RF.csv')

plt.rc('font', family='Times New Roman', size=10)
linewidth = 0.4
plt.rcParams['axes.linewidth'] = linewidth
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['xtick.major.width'] = linewidth
plt.rcParams['ytick.major.size'] = 3
plt.rcParams['ytick.major.width'] = linewidth
fig = plt.figure(figsize=(7.5, 2.3), dpi=150)

cz_name_list = ['Global', 'Tropical', 'Temperate', 'Cold', 'Arid']
plt_id = 1
for cz_name in cz_name_list:
    df_impor_wi_cz = df_impor_wi[df_impor_wi['cz'] == cz_name].reset_index(drop=True)
    x_names = df_impor_wi_cz['variable'][::-1]
    importances = df_impor_wi_cz['importance'][::-1]

    plt.subplot(1, 5, plt_id)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.barh(x_names, importances, color='gray')
    plt.xlabel('')
    if plt_id > 1:
        plt.yticks([0,1,2,3,4,5], [])
    plt_id += 1
plt.tight_layout()
plt.show()
```
